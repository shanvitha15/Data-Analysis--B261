* method for removing bias from evaluating data by  numerical analysis.
* useful for collecting the interpretations of research, developing statistical models, and planning surveys and studies.
* it can be also defined as a science of collecting and analyzing data to identify trends and patterns and presenting them, to make use of data to derive meaningful information.

#### Types:

1. <b>Descriptive Analysis</b> :
 involves collecting, interpreting, analyzing, and summarizing data to present them in the form of charts, graphs, and tables. 
 it simply makes the complex data easy to read and understand.
 
2. <b>Inferential Analysis</b> :
 focuses on drawing meaningful conclusions on the basis of the data analyzed. It studies the relationship between different variables or makes predictions for the whole population.
 
3. <b>Predictive Analysis</b> :
statistical analysis that analyzes data to derive past trends and predict future events on the basis of them

4. <b>Prescriptive Analysis</b> :
analysis of data and prescribes the best course of action based on the results, that helps you make an informed decision. 
5. <b>Exploratory Data Analysis</b> :
similar to inferential analysis, difference is that it involves exploring the unknown data associations.
6. <b>Causal Analysis</b> :
determines why something happens and its effect on other variables.

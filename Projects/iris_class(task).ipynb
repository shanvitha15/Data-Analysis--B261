### Code for Statistical Measures: 
#### Write a class which takes the data set as argument and write methods for the following statistical measures without using built-in functions:of Mean Deviation 

* Arithmetic Mean
* Geometric Mean
* Harmonic Mean
* Mode
* Median
* Describe
* Variance 
* standard deviation
* max 
* min 
* range 
* IQR 
* Quartiles 
* Co-efficient of Range
* Co-efficient of Variation
* Co-efficient of Standard Deviation
* Co-efficient of Quartile Deviation
* Co-efficient of Mean Deviation

```python
class statistics:
    def __init__(self, data):
        self.data = data

    def arithmetic_mean(self):
        total_sum = np.sum(self.data)
        return total_sum / len(self.data)

    def geometric_mean(self):
        total_product = np.prod(self.data)
        return total_product ** (1/len(self.data))

    def harmonic_mean(self):
        total_sum = 0
        for value in self.data:
            total_sum += 1 / value
        return len(self.data) / total_sum

    def mode(self):
        values, counts = np.unique(self.data, return_counts=True)
        return values[counts == np.max(counts)]

    def median(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        if n%2==0:
            return(sorted_data[n//2]+sorted_data[n//2-1])/2
        else:
            median=sorted_data[n//2]
        return median

    def describe_data(self):
        n = len(self.data)
        mean = sum(self.data) / n
        variance = sum((x - mean) ** 2 for x in self.data) / (n - 1)
        std_deviation = variance ** 0.5
        return n, mean, variance, std_deviation, min(self.data), max(self.data)

    def variance(self):
        mean = self.arithmetic_mean()
        variance = sum((self.data - mean) ** 2) / len(self.data)
        return variance

    def standard_deviation(self):
        return self.variance() ** 0.5

    def max(self):
        return max(self.data)
    
    def min(self):
        return min(self.data)

    def range(self):
        return self.max() - self.min()

    def iqr(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        q1 = sorted_data[n // 4]
        q3 = sorted_data[(3 * n) // 4]
        return q3 - q1

    def quartiles(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        q1 = sorted_data[n // 4]
        q2 = sorted_data[n // 2]
        q3 = sorted_data[(3 * n) // 4]
        return q1, q2, q3

    def coefficient_of_range(self):
        return self.range() / self.iqr()
    
    def coefficient_of_variation(self):
        return (self.standard_deviation() / self.arithmetic_mean()) * 100
    
    def coefficient_of_standard_deviation(self):
        return self.standard_deviation() / self.arithmetic_mean()
    
    def coefficient_of_quartile_deviation(self):
        return self.iqr() / self.arithmetic_mean()
```

```python
df = pd.read_csv(r"C:\Users\HP PC\OneDrive\Desktop\Iris.csv")
```

```python
col_name = df['SepalLengthCm']
stats = statistics(col_name)
```

```python
print("Arithmetic Mean:", stats.arithmetic_mean())
print("Geometric Mean:", stats.geometric_mean())
print("Harmonic Mean:", stats.harmonic_mean())
print("Mode:", stats.mode())
print("Median:", stats.median())
print("Describe:", stats.describe_data())
print("Variance:", stats.variance())
print("Standard Deviation:", stats.standard_deviation())
print("Max:", stats.max())
print("Min:", stats.min())
print("Range:", stats.range())
print("IQR:", stats.iqr())
print("Quartiles:", stats.quartiles())
print("Coefficient of Range:", stats.coefficient_of_range())
print("Coefficient of Variation:", stats.coefficient_of_variation())
print("Coefficient of Standard Deviation:", stats.coefficient_of_standard_deviation())
print("Coefficient of Quartile Deviation:", stats.coefficient_of_quartile_deviation())
```

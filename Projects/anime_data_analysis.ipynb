### 5. Is there a connection between the years of an anime and its rating type?

```python
anime['start_Date'] = pd.to_datetime(anime['start_Date'])
anime['Year'] = anime['start_Date'].dt.year
year_rating_type = pd.crosstab(index=[anime['Year'], anime['Rating_type']], columns=anime['Description'], margins=True)
year_rating_type = year_rating_type.drop('All', axis=1)
year_rating_type = year_rating_type.drop('All', axis=0)
```

```python
year_rating_type
```

```python
custom_colors = ['#2C51A2', '#536FA9', '#6579B2', '#849ACB', '#98ABC9', '#A7B7D7', '#ABBCDD']
def apply_custom_colors(val):
    if val == '17+ (violence & profanity)':
        return 'background-color: #536FA9'  
    else:
        return f'background-color: {custom_colors[val % len(custom_colors)]}' 
styled_df = year_rating_type.style.applymap(apply_custom_colors)
styled_df
```

```python
import math
custom_colors = ['#2C51A2', '#536FA9', '#6579B2', '#849ACB', '#98ABC9', '#A7B7D7', '#ABBCDD']
years = year_rating_type.index.get_level_values(0).unique()
num_plots = len(years)
num_cols = 3
num_rows = math.ceil(num_plots / num_cols)
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(18, 6*num_rows))
axes = axes.flatten()
for i, year in enumerate(years):
    year_data = year_rating_type.loc[year]
    ax = year_data.plot(kind='bar', stacked=True, color=custom_colors, ax=axes[i])
    ax.set_title(f'Year {year} Rating Type Distribution')
    ax.set_xlabel('Rating Type')
    ax.set_ylabel('Count')
    ax.legend(title='Description')
    ax.set_xticks(range(len(year_data.index)))
    ax.set_xticklabels(year_data.index)
for j in range(num_plots, num_rows * num_cols):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
```
